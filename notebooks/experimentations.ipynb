## Experimenting with getting YouTube URLS

```python
import yt_dlp

def get_latest_videos(channel_url, limit=5):
    ydl_opts = {
        "quiet": True,
        "extract_flat": True,
        "playlistend": limit,
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        result = ydl.extract_info(f"{channel_url}/videos", download=False)

    return result["entries"][:limit]
```

```python
get_latest_videos("https://www.youtube.com/@FPLHarry")
```

```python
def get_latest_video_urls(channel_url, limit=5):
    videos = get_latest_videos(channel_url, limit)
    return [video["url"] for video in videos]
```

## FPL Api

```python
# src/services/fpl_service.py
from __future__ import annotations

import requests
from typing import Any


class FPLService:
    BASE_URL = "https://fantasy.premierleague.com/api"

    def __init__(self, timeout: int = 20) -> None:
        self.timeout = timeout
        self.session = requests.Session()
        self.session.headers.update(
            {
                "User-Agent": "fpl-agent/0.1",
                "Accept": "application/json",
            }
        )

    def _get(self, path: str) -> Any:
        url = f"{self.BASE_URL}{path}"
        response = self.session.get(url, timeout=self.timeout)
        response.raise_for_status()
        return response.json()

    def get_bootstrap_static(self) -> dict[str, Any]:
        return self._get("/bootstrap-static/")

    def get_fixtures(self) -> list[dict[str, Any]]:
        return self._get("/fixtures/")

    def get_event_live(self, gameweek: int) -> dict[str, Any]:
        return self._get(f"/event/{gameweek}/live/")

    def get_element_summary(self, player_id: int) -> dict[str, Any]:
        return self._get(f"/element-summary/{player_id}/")

    def get_entry(self, team_id: int) -> dict[str, Any]:
        return self._get(f"/entry/{team_id}/")

    def get_entry_picks(self, team_id: int, gameweek: int) -> dict[str, Any]:
        return self._get(f"/entry/{team_id}/event/{gameweek}/picks/")
```

## Parallalezation Agentic Workflow with FL Experts

We do not need to create different agents for every transcription. The best efeecient workflow wuld be to create one agent that we can feed all the different transcriptions and then have one aggregator function to make a final decision based on the structured outputs.

```python
import asyncio
from agents import Agent, Runner, trace
from pydantic import BaseModel
from typing import List


class ExpertAnalysis(BaseModel):
    expert_name: str
    video_title: str
    strategy_summary: List[str]
    fixture_insights: List[str]
    player_picks: List[str]
    transfer_plan: List[str]
    captaincy: List[str]
    chip_strategy: List[str]
    risks: List[str]
    final_takeaways: List[str]


fpl_analysis_agent = Agent(
    name="FPL Transcript Analyst",
    instructions=instructions,
    output_type=ExpertAnalysis,
)
```

Transcription job 

```python
async def analyze_transcripts(transcript_jobs: list[dict]) -> list[ExpertAnalysis]:
    with trace("Parallel FPL transcript analysis"):
        results = await asyncio.gather(
            *[
                Runner.run(
                    fpl_analysis_agent,
                    f"""
                    Expert: {job['expert_name']}
                    Video title: {job['video_title']}
                    Gameweek: {job['gameweek']}

                    Transcript:
                    {job['transcript']}
                    """
                )
                for job in transcript_jobs
            ]
        )

    return [result.final_output for result in results]
```

```python
outputs = await analyze_transcripts(transcript_jobs)

for output in outputs:
    print(output.model_dump())
    print("\n")
```
